```python
df_sales = pd.read_csv('일별매출_2021-01-02~2021-07-20.csv')
df_sales.index = df_sales['Unnamed: 0']
df_sales = df_sales[['무보까국밥(장기점)', '무보까국밥(성서점)', '무보까국밥(상인점)', '요일']]
df_sales.columns = ['장기점', '성서점', '상인점', '요일']
df_sales['용산점'] = df_ys['무보까국밥(용산점)']
df_sales
```

```python
df_total = pd.DataFrame()
df_total['달서'] = df_cases['달서']
df_total['장기점'] = df_sales['장기점']
df_total['성서점'] = df_sales['성서점']
df_total['상인점'] = df_sales['상인점']
df_total['용산점'] = df_sales['용산점']
df_total['요일'] = df_sales['요일']
df_total
```

```python
#del df_total(lambda x : x = '월', df_total['요일'])
df_total['요일']
```

```python
df_total = df_total.reset_index()
df_total['Unnamed: 0'] = pd.to_datetime(df_total['Unnamed: 0'])
df_total['요일_번호'] = df_total['Unnamed: 0'].dt.weekday
#del df_total['index']
df_total
```

```python
df_total.index = df_total['Unnamed: 0']
del df_total['Unnamed: 0']
df_total
```

```python
df_total
```

```python
d_list = []
j_list = []
s_list = []
si_list = []
ys_list = []
i_list = []
for a in range(0, len(df_total)) :
    if df_total.iloc[a, 6] == 0 or df_total.iloc[a, 6] == 1 or df_total.iloc[a, 6] == 2 or df_total.iloc[a, 6] == 3 or df_total.iloc[a, 6] == 4 :
        d_list.append(df_total.iloc[a, 0])
        j_list.append(df_total.iloc[a, 1])
        s_list.append(df_total.iloc[a, 2])
        si_list.append(df_total.iloc[a, 3])
        ys_list.append(df_total.iloc[a, 4])
        i_list.append(df_total.index[a])
df_week = pd.DataFrame()
df_week['달서'] = d_list
df_week['장기'] = j_list
df_week['성서'] = s_list
df_week['상인'] = si_list
df_week['용산'] = ys_list
df_week.index = i_list
df_week
```

```python
df_week['장기'] = [int(j.replace(',', '')) for j in df_week['장기']]
df_week['성서'] = [int(k.replace(',', '')) for k in df_week['성서']]
df_week['상인'] = [int(l.replace(',', '')) for l in df_week['상인']]
df_week['용산'] = [int(m.replace(',', '')) for m in df_week['용산']]
df_week
```

```python
df_week.iloc[40:]
### 41 -> 3월 1일
df_week_3 = df_week.iloc[40:]
df_week_3
```
