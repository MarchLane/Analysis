```python
import os
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
```

```python
os.chdir(r'G:\공유 드라이브\SicParvisMagna\데이터\코로나_매출')
df_covid = pd.read_csv('대구구별_확진자수_21-01-02~21-07-20.csv')
df_covid
```

```python
df_sales = pd.read_csv('일별매출_2021-01-02~2021-07-20.csv')
df_sales
```

## 달서구 확진자 수 VS 장기점 매출

```python
dalseo_jangki = pd.DataFrame()
dalseo_jangki['covid_dalseo'] = df_covid['달서']
dalseo_jangki['sales_jangki'] = [int(a.replace(',', '')) for a in df_sales['무보까국밥(장기점)']]
dalseo_jangki
```

```python
scaler = StandardScaler()
dalseo_jangki_sc = scaler.fit_transform(dalseo_jangki)
dalseo_jangki_sc = pd.DataFrame(dalseo_jangki_sc)
dalseo_jangki_sc.columns = ('covid_dalseo', 'sales_jangki')
dalseo_jangki_sc
```

```python
train_features, test_features, train_labels, test_labels = train_test_split(dalseo_jangki_sc['covid_dalseo'], dalseo_jangki_sc['sales_jangki'])
```

```python
test_features
```

```python
train_labels
```

```python
model = LogisticRegression()
model.fit(train_features, train_labels)
```

```python
model.score(train_features, train_labels)
```
