```python
import os
import pandas as pd
import numpy as np
import statsmodels.api as sm
from statsmodels.tsa.api import VAR
```

#### 제가 창업한 기업에서 실제로 사용 중인 기업의 매출 데이터 입니다.

```python
os.chdir(r'G:\공유 드라이브\SicParvisMagna\데이터\인공지능모델_데이터셋구축\매출데이터')
df_sales = pd.read_csv('무보까 일별 매출_모든 매장_13-10-02_21-03-31.csv')
df_sales.index = df_sales['Unnamed: 0']
del df_sales['Unnamed: 0']
df_sales
```

#### 여러 지점 중에 하나의 지점만 선택하여 분석할 예정입니다.

```python
df_si = pd.DataFrame(df_sales['상인']).dropna()
df_si = df_si[2:]
df_si
```

#### 일별 매출과 KOSPI, 금리 간의 상관관계를 분석하고 회귀분석을 통해 일별 매출을 예측해보려고 합니다.

```python
os.chdir(r'G:\공유 드라이브\SicParvisMagna\데이터\인공지능모델_데이터셋구축\경제지표')
df_koint = pd.read_csv('KOSPI_금리_휴일메꿈_2014-2021.csv', encoding = 'cp949')
df_koint.index = df_koint['Unnamed: 0']
del df_koint['Unnamed: 0']
df_koint
```

```python
df_si['KOSPI'] = df_koint['KOSPI']
df_si['금리'] = df_koint['금리']
df_si
```

#### 데이터 안정화를 위해 log 를 취하거나 log 를 취하고 차분을 해줍니다.

```python
df_si_log = np.log(df_si)
df_si_log = df_si_log[:-120]
df_si_log
```

```python
df_si_logdiff = np.log(df_si).diff().dropna()
df_si_logdiff = df_si_logdiff[:-120]
df_si_logdiff
```

#### 모델은 VAR 모델을 사용하였고, aic 를 기준으로 lag order 를 설정하였습니다.

```python
model = VAR(df_si_logdiff)
results_all = model.fit(maxlags = 100, ic = 'aic')
results_all.summary()
```

#### VAR 모델을 통해 나온 식으로 120일 까지를 예측해 보았습니다.

```python
np_forc_120 = results_all.forecast(df_si_logdiff.values[-results_all.k_ar : ], 120)
np_forc_120
```

```python
df_forc_120 = pd.DataFrame(np_forc_120)
df_forc_120.columns = ['상인', 'KOSPI', '금리']
df_forc_120
```

```python
14.460780	7.876354	-0.072571
```

#### log 와 차분을 해줬기 때문에 다시 데이터를 원본으로 돌리는 과정을 거칩니다.

```python
sales = 14.460780
sales_list = []
for c in df_forc_120['상인'] :
    sales += c
    sales_list.append(sales)

kos = 7.876354
kos_list = []
for b in df_forc_120['KOSPI'] :
    kos += b
    kos_list.append(kos)
    
intr = -0.072571
intr_list = []
for a in df_forc_120['금리'] :
    intr += a
    intr_list.append(intr)
```

```python
results_forc_120 = pd.DataFrame()
results_forc_120['상인'] = sales_list
results_forc_120['KOSPI'] = kos_list
results_forc_120['금리'] = intr_list
results_forc_120
```

```python
np.exp(results_forc_120)
```

#### 실제값과 예측값을 비교합니다.

```python
results_120 = pd.DataFrame(df_si['상인'][-120:])
results_120 = results_120.reset_index()
results_120['예측값'] = np.exp(results_forc_120)['상인']
results_120.index = results_120['Unnamed: 0']
del results_120['Unnamed: 0']
results_120
```

#### 120일을 예측한 결과 

```python
results_120['오차제곱근'] = [((a - b)**2)**0.5 for a, b in zip(results_120['상인'], results_120['예측값'])]
results_120['오차비율'] = [d / c for c, d in zip(results_120['상인'], results_120['오차제곱근'])]
results_120
```

```python
os.chdir(r'C:\Users\alswo\Desktop\4학년\빅데이터')
results_120.to_csv('상인_120일_매출예측.csv', encoding = 'cp949')
```

#### '오차제곱근 / 실제매출' 방식으로 오차비율을 구했고, 120일의 평균은 약 58.8%의 비율을 보이고 있기 때문에 성능이 아주 떨어지는 모델이라는 결론을 갖게 되었습니다. 아무래도 120일이라는 기간이 너무 길다 보니 이런 결과가 나온것 같은데, 처음 5일 정도를 예측했을 때는 10%대의 오차비율을 가졌습니다. 그리고 KOSPI 와 시장금리가 매출을 예측하는데에 크게 도움이 되지 않아서 모델 성능이 안좋을 수도 있겠다는 생각이 들었습니다. 그래서 기온, 습도, 풍속과 같은 날씨 데이터를 변수로 추가할 계획입니다. 또 한가지는 예측하는 일수를 조금 줄여볼 생각입니다.

#### 원래는 이 VAR 모델의 회귀식을 이용해서 예측을 하려던 것이 아니었습니다. VAR 모델로는 매출이라는 변수와 Coefficient 가 가장 높은 시차를 가진 변수를 LSTM 에 학습시킬 때 활용하려고 했었습니다. 예를 들어 t 시간의 매출과 가장 Coefficient 가 높은 시간대의 변수는 KOSPI(t-3) 이고 시장금리(t-5) 라고 가정했을 때, LSTM 학습 시계열 데이터를 KOSPI는 3만큼의 시차를 두고 시장금리는 5만큼의 시차를 둔 데이터를 학습시키려고 하고있습니다. 이와 관련해서 교수님의 의견이 궁금합니다.
