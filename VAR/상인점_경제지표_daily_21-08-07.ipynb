```python
import os
import pandas as pd
import numpy as np
import statsmodels.api as sm
from statsmodels.tsa.api import VAR
```

```python
os.chdir(r'G:\공유 드라이브\SicParvisMagna\데이터\인공지능모델_데이터셋구축\VAR_전처리')
df_var = pd.read_csv('매출금액_시장금리_kospi_휴일채움.csv', encoding = 'cp949')
df_var.index = df_var['매출일자']
del df_var['매출일자']
df_var = df_var[:-119]
df_var
```

```python
df_var['매출금액'] = [int(v.replace(',', '')) for v in df_var['매출금액']]
df_var
```

```python
df_var_log_diff = np.log(df_var).diff().dropna()
df_var_log_diff
```

```python
var = df_var_log_diff[['매출금액', '시장금리']]
var
```

```python
model = VAR(var)
results_all = model.fit(maxlags = 100, ic = 'aic')
results_all.summary()
```
